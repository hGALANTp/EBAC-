B) Exclua nessa etapa as linhas das colunas com dados faltantes que você analisar que esse método é o ideal. Justifique o motivo de estar excluindo essas linhas.

```python
# Aqui optei por excluir pois se tratava de poucos dados nulos.

df.dropna(subset=['Genero'], inplace=True)
df.info()


```

C) Nessa etapa substitua pela média, mediana ou moda as linhas das colunas com valores faltantes que você analisou e entendeu que esse era o melhor método. Não se esqueça de justificar sua escolha.

```python
# Para preservar o maximo de dados, aqui optei preencher com a media os dados faltantes. foi observado que a media e mediana estavam proximas não conteno outliers relevantes.

media_pgto = df['Pagamento_Mensal'].mean()
df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(media_pgto)


```

**Dica variáveis numéricas:** Não se esqueça de verificar a distrbuição dos dados, média e mediana.

**Dica variáveis categóricas:** Caso precise substituir uma variável categórica utilize um gráfico de barras para verificar a que mais aparece e utilizar ela na substituição.
Ou você pode optar por excluir a coluna, mas justifique sua escolha.

Caso opte por substituir você pode utilizar a função fillna:
df['SUACOLUNA'].fillna("VALOR QUE SUBSTITUIRA O NULO", inplace=True)
Ela funcionará como a replace.

```python
# Aqui optei em preencher como não, interpretando que quem nao preencheu não estaria usando esse tipo de serviço. 

df['PhoneService'] = df['PhoneService'].fillna('No')
```

# 3 - Verifique se encontramos valores digitados incorretamente, ou com letras maiusculas ou minusculas, ou até mesmo mesmos valores porém escritos de forma dirente e corrija.

```python
#seu código aqui
#Aqui só vamos pegar as colunas categoricas. 

colunas_texto = df.select_dtypes(include='object').columns
print(len(colunas_texto)) # coloquei pra verificar se os prints no final seria na mesma quantidade de colunas capturadas. 
for col in colunas_texto:
    print(f"\nColuna: {col}")
    print("Valores únicos :")
    print(df[col].value_counts())



```

```python
#Aqui percebi que o PhoneService existe valores nulos e por isso ele não trás os valores unicos pra gente conferir, sendo assim vamos tratar ele tbm. 
df['PhoneService'] = df['PhoneService'].str.strip().fillna('No')

df['Genero'] = df['Genero'].str.lower().map({'f':'Female', 'm':'Male'})
df['Servico_Internet'] = df['Servico_Internet'].replace('dsl','DSL')

colunas_texto = df.select_dtypes(include='object').columns
print(len(colunas_texto)) # coloquei pra verificar se a mesma quantidade 
for col in colunas_texto:
    print(f"\nColuna: {col}")
    print("Valores únicos :")
    print(df[col].value_counts())
```
